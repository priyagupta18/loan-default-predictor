``` Findings ```
- Less people have applied for more than 15 loans. So, let's combine more than 15 to make a category
- Less people have Completed for more than 15 loans. So, let's combine more than 15 to make a category
- Less people have active for more than 5 loans. So, let's combine more than 5 to make a category

````python
def clubbing(col, threshold):
    if col>=threshold:
        return threshold
    else:
        return col
        
df['total_tl'] = df['total_tl'].apply(lambda x:clubbing(x,threshold=15))
df['complete_tl'] = df['total_tl'].apply(lambda x:clubbing(x,threshold=15))
df['active_tl'] = df['total_tl'].apply(lambda x:clubbing(x,threshold=5))
````

### Outlier detection

````python
boxplot_cols = ['total_tl', 'complete_tl',
       'active_tl', 'amount_overdue_sum', 'loan_amount_avg', 'loan_amount_min',
       'loan_amount_max', 'num_times_delinquent',
       'months_since_recent_delinquency', 'automobile_loans', 'business_loans',
       'consumer_loans', 'mortgages_and_real_estate_loans', 'other_loans',
       'enquiry_count', 'max_enquiry_amount', 'time_since_last_enquiry',
       'business_loans_enquiries', 'consumer_loans_enquiries',
       'revolving_loans_enquiries',
       'mortgages_and_real_estate_loans_enquiries', 'other_loans_enquiries',
       'automobile_loans_enquiries']


# Loop through the list of columns and plot each one
for i, col in enumerate(boxplot_cols):
    plt.figure(figsize=(6, 4)) 
    sns.boxplot(data=df, x=col)
    plt.title(f'Box Plot for {col}')
    plt.xlabel('')
    plt.show()
````

````python
#outlier detection
outlier_cols = ['loan_amount_avg','loan_amount_min','loan_amount_max',
                'num_times_delinquent','consumer_loans',
                'enquiry_count','max_enquiry_amount',
               'time_since_last_enquiry','business_loans_enquiries','consumer_loans_enquiries',
                'revolving_loans_enquiries','mortgages_and_real_estate_loans_enquiries',
               'other_loans_enquiries','automobile_loans_enquiries']

def cap_outliers_iqr(df, columns):
    for column in columns:
        if column in df.columns:
            # Calculate Q1 (25th percentile) and Q3 (75th percentile)
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            
            # Calculate IQR
            IQR = Q3 - Q1
            
            # Determine outlier bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Cap the outliers
            df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    
    return df

df_clean = cap_outliers_iqr(df, outlier_cols)
````

````python
for i, col in enumerate(boxplot_cols):
    plt.figure(figsize=(6, 4)) 
    sns.boxplot(data=df_clean, x=col)
    plt.title(f'Box Plot for {col}')
    plt.xlabel('')
    plt.show()
````
